```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):

  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

# Section 04 : Merging the Datasets

```python
from datetime import datetime
```

```python
import pandas as pd
```

```python
def convert_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy


```

```python
sales_df = convert_hourly(sales_df, 'timestamp')
stock_df = convert_hourly(stock_df, 'timestamp')
temp_df = convert_hourly(temp_df, 'timestamp')
```

```python
sales_df.head()
```

```python
stock_df.head()
```

```python
temp_df.head()
```

```python
sales_comb= sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_comb.head()
```

```python
stock_comb = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_comb.head()
```

```python
temp_comb = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_comb.head()

```

```python

merged_df = stock_comb.merge(sales_comb, on=['timestamp', 'product_id'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.merge(temp_comb, on='timestamp', how='left')
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

```python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates()

product_price = sales_df[['product_id', 'unit_price']]
product_price = product_price.drop_duplicates()

merged_df = merged_df.merge(product_categories, on="product_id", how="left")
merged_df.head()
```

```python

merged_df = merged_df.merge(product_price, on="product_id", how="left")
merged_df.head()
```

## Section 05 : **Feature Engineering**

```python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.drop(columns=['product_id'], inplace=True)
merged_df.head()
```

## Section 06: **Modelling**

```python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
K = 10
split = 0.75
!pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
accuracy = []

for fold in range(0, K):

  model = RandomForestRegressor()
  scaler = StandardScaler()

  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)
  trained_model = model.fit(X_train, y_train)

  y_pred = trained_model.predict(X_test)

  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)
  print(f"TRY {fold + 1}: Error = {mae:.3f}")

print(f"Average Absolute Error: {(sum(accuracy) / len(accuracy)):.2f}")

```
